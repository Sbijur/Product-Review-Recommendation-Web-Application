```python
def web_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--verbose")
    options.add_argument('--no-sandbox')
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument("--window-size=1920, 1200")
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    return driver
```

```python
import time
```

```python
from tqdm import tqdm
```

```python
titles = []
images = []
productLink = []
ratings = []
prices = []
categories = []  # Added a list to store categories
for i in tqdm(range(No_of_pages)):
    url = 'https://www.amazon.in/s?k=' + search_term + '&page='+str(i+1)
    driver.get(url)
    time.sleep(3)
    html_data = BeautifulSoup(driver.page_source, 'html.parser')
    products = html_data.find_all('div',{'data-component-type':'s-search-result'})
    for product in products:

        title = product.find('span',{'class':'a-size-medium a-color-base a-text-normal'})
        if title == None:
            title = "NA"
        else:
            title = product.find('span',{'class':'a-size-medium a-color-base a-text-normal'}).text
        titles.append(title)

        img = product.find('img')['src']
        images.append(img)

        rating = product.find('span',{'class':'a-icon-alt'})
        if rating  == None:
            rating = 'No Rating'
        else:
            rating = product.find('span',{'class':'a-icon-alt'}).text
        ratings.append(rating)

        price = product.find('span',{'class':'a-price-whole'})
        if price == None:
            price = 'No Price'
        else:
            price = 'Rs ' + product.find('span',{'class':'a-price-whole'}).text
        prices.append(price)

        p1 = product.find('a')['href']
        p2 = "https://www.amazon.in"+p1
        decoded_url = html.unescape(p2)
        productLink.append(decoded_url)


        # Extracting category information
        category = category_name
        categories.append(category)
```

```python
len(titles)
```

```python
data = pd.DataFrame({'titles':titles, 'images':images,'ratings':ratings, 'prices':prices, 'categories':categories, 'link':productLink})
```
